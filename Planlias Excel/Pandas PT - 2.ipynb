```python
import pandas as pd
```

```python
df1 = pd.read_excel("Aracaju.xlsx")
df2 = pd.read_excel("Fortaleza.xlsx")
df3 = pd.read_excel("Natal.xlsx")
df4 = pd.read_excel("Recife.xlsx")
df5 = pd.read_excel("Salvador.xlsx")
```

```python
df = pd.concat([df1, df2, df3, df4, df5])
```

```python
# retorna as ultimas 5 linhas
df.tail()
```

```python
# verifcar o tipo de dado de cada coluna
df.dtypes
```

```python
# alterar o tipo de dado de uma coluna 
df['LojaID'] = df['LojaID'].astype('object')
```

```python
# verifcar o tipo de dado de cada coluna
df.dtypes
```

```python
# retorna as primeiras 5 linhas
df.head()
```

```python
#TRATANDO VALORES FALTANTES
```

```python
#consultando linhas com valores faltantes e soma todas as faltantes
df.isnull().sum()
```

```python
# subistituindo valores nulos
df["Vendas"] = df["Vendas"].fillna(df["Vendas"].mean())
```

```python
#consultando linhas com valores faltantes e já a media dos faltantes
df.isnull().mean()
```

```python
# retorna linha(s) aleatorias do dataFrame, a quantidade fica em ( )
df.sample(15)
```

```python
# fazendo a media de vendas
df["Vendas"].mean()
```

```python
#subistituindo valores nulos por 0
df["Vendas"] = df["Vendas"].fillna(0)
```

```python
#apagando colunas com valores nulos
df.dropna(inplace= True)
```

```python
#apagando colunas com valores nulos com base em apenas uma coluna
df.dropna(subset=['Vendas'], inplace=True)
```

```python
#apagando colunas com valores nulos com base em todas as colunas
df.dropna(how='all', inplace=True)
```

```python
#Criandos colunas novas
```

```python
#Criando a coluna de receita 
df["Receita"] = df["Vendas"].mul(df["Qtde"])
```

```python
# retorna as primeiras 5 linhas
df.head(50)
```

```python
# fazendo a media de Receita
df["Receita"].mean()
```

```python
# se tivesse a coluna Qtde, esta eéuma forma de conseguir gerar o mesmo valor dela 
df["Receita/Vemdas"] = df["Receita"] / df["Vendas"]
```

```python
# retorna as primeiras 5 linhas
df["Receita/Vemdas"].head()
```

```python
# retorna a maior receita
df["Receita"].max()
```

```python
#retorna a menor recieta
df["Receita"].min()
```

```python
#retorna nn(o numero que tiver dentro) linhas das maiores vendas
df.nlargest(3, "Receita")
```

```python
#retorna nn(o numero que tiver dentro) linhas das menores vendas
df.nsmallest(3, "Receita")
```

```python
#agrupamento por Cidade - ordenando pela soma da receita de cada cidade
df.groupby("Cidade")["Receita"].sum()
```

```python
#ordenando pelo maior valor de receita e retorna os primeiros 10
df.sort_values("Receita", ascending=False).head(10)
```

```python
#Trabalhando com Datas
```

```python
#alterando o formato da coluna data para Inteito
df["Data"] = df["Data"].astype("int64")
```

```python
df.dtypes
```

```python
#alterando o formato da coluna data para Datetime
df["Data"] = pd.to_datetime(df["Data"])
```

```python
df.dtypes
```

```python
# agrupando po data
df.groupby(df["Data"].dt.year)["Receita"].sum()
```

```python
#criando a coluna do ano da Venda
df["Vendas_ano"] = df["Data"].dt.year
```

```python
df.sample(5)
```

```python
# criando as colunas de dia e mes da separados um do outro
df["mes_venda"],df["dia_neda"] = df["Data"].dt.month, df["Data"].dt.day 
```

```python
df.sample(5)
```

```python
#Retorna a data mais antiga
df["Data"].min()
```

```python
#retorna a diferença entre as datas e a menor data do Data frame
df["diferenas_dias"] = df["Data"] - df["Data"].min()
```

```python
df.sample(5)
```

```python
#crianbdo a coluna do trimestre
df["Trimeste_vendas"] = df["Data"].dt.quarter
```

```python
df.sample(5)
```

```python
#criando uma variavel para busca todas as vendas da da de de  2019 e mes 3(março)
vendas_marco_29 = df.loc[(df["Data"].dt.year == 2019) & (df["Data"].dt.month == 3)]
```
